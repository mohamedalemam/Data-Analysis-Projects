## Importing Libraries & Methods

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

## Inputting the URL

```python
url = "https://wuzzuf.net/search/jobs/?q=Business+Analyst&a=navbl"
```

## Create a Client-based Request to Get the URL

```python
client = urlopen(url)
```

## Getting the HTML Code of the Full Page

```python
html = client.read()
```

```python
html
```

## Creating an HTML Parser Using Beautifu/Soup

```python
soup = bs(html, "html.parser")
```

```python
soup
```

## Creating a Container for the Needed Data

```python
containers = soup.find_all("div", {"class" : "css-1gatmva e1v1l3u10"})
```

## Bringing it All Together

```python
f = open("Wuzzaf_Scraping.csv", mode = "w")
header = "job_title, company_name, job_type\n"
f.write(header)
```

```python
for x in containers:
    jtitle = x.findAll("h2", {"class" : "css-m604qf"})
    job_title =  jtitle[0].text.strip()
    
    cname = x.findAll("a" , {"class" : "css-17s97q8"})
    compant_name =  cname[0].text.strip()
    
    jtype = x.findAll("span" , {"class" : "css-1ve4b75 eoyjyou0"})
    job_type = jtype[0].text.strip()
    
#     print(job_title)
#     print(compant_name)
#     print(job_type)
#     print()

#     print(job_title+ ', ' + compant_name + ', ' + job_type + "\n")
#     print()
    f.write(job_title+ ', ' + compant_name + ', ' + job_type + "\n")
f.close()
```
